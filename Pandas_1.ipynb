### GroupBy function

```python
income.groupby("Index")["Year2005", "Year2009"].sum()
```

```python
# use agg( ) function which is used to aggregate the data. 

income.groupby('Index')["Year2002", 'Year2003'].agg(["min", "max", "mean"])
```

```python
# finds minimum and maximum values for Year2008 and only mean for Y2012

income.groupby('Index').agg({"Year2008": ['min', "max"], "Year2015": "mean"})
```

```python
#  to rename the columns after groupby, you can use tuple.

income.groupby('Index').agg({"Year2008": [("Year2008_min", 'min'), ("Year2008_max", "max")],
                             "Year2012": [("Year2012_mean", "mean")]})
```

```python
dt = income.groupby("Index").agg({"Year2004": ['min', "max"], "Year2009": "mean"})
dt.columns = ["Year2004_min", "Year2004_max", "Year2009_mean"]
dt
```

```python
# Groupby more than 1 column

income.groupby(["Index","State"]).agg({"Year2004": ["min", "max"], "Year2007": "mean"})
```

```python
# By default, option as_index=True is enabled in groupby
# which means the columns you use in groupby will become an index in the new dataframe.
# To disable it, you can make it False which stores the variables you use in groupby
# in different columns in the new dataframe.

income.groupby(["Index", "State"], as_index=False)["Year2005", "Year2007"].sum()
```

### Pivot Table

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

* how both gender and class affected survival

```python
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
```

* Here is the equivalent to the preceding operation using the pivot_table method of
DataFrames:

```python
titanic.pivot_table('survived', index='sex', columns='class')
```

### Multilevel pivot tables

we might be interested in looking at age as a third dimension. We’ll bin the age using the pd.cut function:

```python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
```

* We can apply this same strategy when working with the columns as well; let’s add info
on the fare paid using pd.qcut to automatically compute quantiles:

```python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
```
