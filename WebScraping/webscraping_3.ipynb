# WEB SCRAPING

```
import requests
from requests import get
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np

from time import sleep
from random import randint
```

```
#initialize empty lists where you'll store your data

abstracts=[]
titles = []
years=[]
journals = []
```

```
url_base = "https://www.tandfonline.com/toc/rtrr20/"
```

```
vol = range(36,40)
issue = range(1,4)

urls = [url_base + str(v) + "/"+ str(i) + "?nav=tocList" for v in vol for i in issue]
```

```
urls
```

```
for url in urls:
  
  dois= []
  
  results = requests.get(url)
  
  soup = BeautifulSoup(results.text, "html.parser")
  soup.prettify()
  links = soup.find_all('a', attrs= {"class":"ref nowrap"})

  for link in links:
    dois.append(link['href'])

  url_articles = ["https://www.tandfonline.com" + dois[i] for i in range(len(dois))]

  for url_article in url_articles:
    
    print(url_article)
    
    res = requests.get(url_article)
    soup_article = BeautifulSoup(res.text, "html.parser")

    abstract = soup_article.find('div', attrs= {'class':"abstractSection abstractInFull"}).text if soup_article.find('div', attrs= {'class':"abstractSection abstractInFull"}) else " "
    abstracts.append(abstract)

    title = soup_article.find('span', attrs= {'class':"NLM_article-title hlFld-title"}).text
    titles.append(title)

    year = soup_article.find('div', attrs= {'class':"title-container"}).text if soup_article.find('div', attrs= {'class':"title-container"}) else " "
    years.append(year)

    sleep(randint(2,10)) 
```
