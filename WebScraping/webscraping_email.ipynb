````
import time
import pandas as pd
import numpy as np
import math
````

````
from selenium.webdriver.common.action_chains import ActionChains

from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as ec
````

````
main_url="https://thegoodpubguide.co.uk/pubs/?paged=1&order_by=category&search=pubs&pub_name=&postal_code="
````

````
wd.get(main_url)
````

````
df = pd.DataFrame(columns =["Name", "Url"])
````

````
for i in range(100):

  if i>0:
    if i ==1:
      x=i+1
    elif 5>i>1:
      x=i+2
    else:
      x=7

    element = WebDriverWait(wd, 20).until(ec.element_to_be_clickable((By.XPATH, f'//*[@id="search-results"]/div[2]/div/div/ul/li[{x}]/a')))
    ActionChains(wd).move_to_element(element).click().perform()

    time.sleep(5)
  
  names = wd.find_elements_by_class_name('pub-search-results-title')
  names_list = [names[i].text for i in range(len(names))]

  in_urls = [wd.find_element_by_xpath(f'//*[@id="search-results"]/div[1]/div[{i}]/div/div/div[2]/div[1]/p/a').get_attribute("href") for i in range(1,21)]

  df = df.append(pd.DataFrame({"Name":names_list, "Url": in_urls}), ignore_index=True)
````
