```
# Import a scrapy Selector, response
import scrapy
from scrapy import Selector
```

```
url = 'https://www.tripadvisor.com.tr/Restaurants-g297977-Bursa.html'
```

```
# Import requests
import requests

# Create the string html containing the HTML source
html = requests.get(url).content

# Create the Selector object sel from html
sel = Selector( text = html )

# Print out the number of elements in the HTML document
print( "You have found: ", len( sel.xpath('//*') ) )
```

```
# Import the CrawlerProcess: for running the spider
from scrapy.crawler import CrawlerProcess
```

```
# Create the Spider class
class Rest_Spider(scrapy.Spider):
  name = "rest_spider"
  # start_requests method
  def start_requests(self):
    yield scrapy.Request(url = url,
                         callback = self.parse_front)
  # First parsing method
  def parse_front(self, response):
    rest_links = response.css('a._15_ydu6b::attr(href)')

    # get links and pass each one in "parse_page" function 
    for link in rest_links:
      url = response.urljoin(link.extract())
      yield scrapy.Request(url, callback = self.parse_page)
    
    # get next page link and pass it in "parse_front" function  
    next_page = response.xpath ("//*[@id='EATERY_LIST_CONTENTS']/div[2]/div/a[contains(@class, 'next')]/@href")
    if next_page:
      url = response.urljoin(next_page.extract_first()) # or next_page[0]
      yield scrapy.Request(url, self.parse_front)

 # Second parsing method
  def parse_page(self, response):
    rest_title = response.css('h1._3a1XQ88S::text').extract_first()
    rest_rating = response.css('svg._3KcXyP0F::attr(aria-label)').extract_first()
    rest_rating_1 = response.css('span.r2Cf69qf::text').extract_first()
    phone = response.css('span._15QfMZ2L > a::text').extract_first()
    rest_dict[rest_title] = [rest_title, rest_rating, rest_rating_1, phone]
```

```
# Run the Spider
process = CrawlerProcess()
process.crawl(Rest_Spider)
process.start()
```
